```python
def get_regular_file_name(filename: str) -> str:
    """Convert a file path to a regular file name by replacing slashes with underscores."""
    return filename.replace("/", "_").replace("\\", "_")
```

```python
pdfs = []
book_files = {}
for book in os.listdir("Books"):
    if book.endswith(".pdf"):
        pdfs.append(f"Books/{book}")
    else:
        book_files[book] = []
        for folder in os.listdir(f"Books/{book}"):
            folder_path = f"Books/{book}/{folder}"
            if os.path.isdir(folder_path):
                for subfile in os.listdir(folder_path):
                    if subfile.endswith((".pdf", ".html", ".json")):
                        book_files[book].append(f"{folder_path}/{subfile}")
print(pdfs)
print(book_files)
```

```python
client_wrapper.list_vector_store_filenames(vector_store_id=vector_store.id)
```

```python
for pdf in pdfs:
    if os.path.basename(pdf) not in client_wrapper.list_vector_store_filenames(vector_store.id):
        file = client_wrapper.get_or_create_file(pdf)
        if not file:
            raise RuntimeError(f"Failed to upload or retrieve file {pdf}")
        vs_file = client_wrapper.get_or_create_vector_store_file(
            vector_store_id=vector_store.id,
            file_id=file.id
        )
        print(f"Uploaded and attached {pdf} with file_id {file.id} to vector store {vector_store.name} as vs_file_id {vs_file.id}")
    else:
        print(f"{pdf} already in vector store {vector_store.name}, skipping upload")
```

```python
filenames = client_wrapper.list_vector_store_filenames(vector_store.id)
for book,bf in book_files.items():
    for file_path in bf:
        if os.path.basename(file_path) not in filenames:
            file = client_wrapper.get_or_create_file(file_path)
            if not file:
                raise RuntimeError(f"Failed to upload or retrieve file {file_path}")
            vs_file = client_wrapper.get_or_create_vector_store_file(
                vector_store_id=vector_store.id,
                file_id=file.id
            )
            print(f"Uploaded and attached {file_path} with file_id {file.id} to vector store {vector_store.name} as vs_file_id {vs_file.id}")
        else:
            print(f"{file_path} already in vector store {vector_store.name}, skipping upload")
```
